```python
## Reading PDF Files

from langchain_community.document_loaders import PyPDFLoader

loader =PyPDFLoader('report.pdf')
docs = loader.load()
docs
```

```python
type(docs)
```

```python
type(docs[0])
```

#### How to recursively split text by charraters
RecursiveCharacterTextSplitter is for Text splitting from Documents

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)

final_document = text_splitter.split_documents(docs)
final_document
```

```python
print(final_document[0])
```

```python
print(final_document[1])
```

Txt file loading and splitting the text

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('speech.txt')

doc = loader.load()
doc
```

```python
## Splitting the text file

speech=""
with open('speech.txt') as f:
    speech = f.read()

txt_splitter = RecursiveCharacterTextSplitter(chunk_size=50, chunk_overlap=10)

final_txt = txt_splitter.create_documents([speech])
print(final_txt[0])
print(final_txt[1])
```

#### How to split by Character Text Splitter
For this CharacterTextSplitter is used

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('speech.txt')

new_doc = loader.load()

new_doc
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(separator="", chunk_size=50, chunk_overlap=10)

text_splitter.split_documents(new_doc)
```

```python
speech=""
with open('speech.txt') as f:
    speech = f.read()

text_splitter = CharacterTextSplitter(separator="", chunk_size=50, chunk_overlap=10)

final_txt = text_splitter.create_documents([speech])
print(final_txt[0])
print(final_txt[1])
```
